# Student Score Prediction

### Importing Required Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

### Load Dataset

```python
data = pd.read_csv('datasets/student_scores.csv')
data.head()
```

```python
X = data['Hours'].values
y = data['Scores'].values

X = X.reshape(-1, 1)
y = y.reshape(-1, 1)

print('X shape:', X.shape, 'y data shape:' , y.shape)
print("No of entries:", X.shape[0])
```

### Visualizing the Dataset

```python
plt.scatter(X, y)
plt.title('Hours Vs Scores')
plt.xlabel('Hours Studied(Hrs)')
plt.ylabel('Scores(out of 100)')
plt.show()
```

### Regression

```python
model = LinearRegression()
model.fit(X, y)

print('Model Score:', model.score(X, y))
```

### Visualizing Regression

```python
# Data Points
plt.scatter(X, y)
plt.title('Hours Vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')

# Regression Line
X_test = [[0], [10]]
plt.plot(X_test, model.predict(X_test), c='orange')

plt.show()
```

### Prediction

What will be the predicted score if a student studies for 9.25hrs/day ?

```python
hrs = [[9.25]]

pred_score = model.predict(hrs)

print("The student's score is predicted to be: ", pred_score.squeeze())
```
